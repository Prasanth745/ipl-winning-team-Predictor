<font size="5">Data Cleaning

```python
#Finding Total Runs Scored in each innings
total_runs = first_dataset.groupby(['match_id','inning']).sum()['total_runs'].add(1).reset_index()
total_runs.head(10)
```

```python
#selecting only first innings to find target score for each match
total_runs = total_runs[total_runs['inning'] == 1]
total_runs
```

```python
#merging total runs df with second dataset
match_df = second_dataset.merge(total_runs[['match_id','total_runs']],left_on='id',right_on='match_id')
```

```python
match_df
```

```python
match_df.columns
```

```python
match_df = match_df[['match_id','city','winner','total_runs']]
```

```python
match_df = match_df.rename(columns={'id':'match_id'})
```

```python
second_innings = match_df.merge(first_dataset,on='match_id')
second_innings
```

```python
second_innings = second_innings[second_innings['inning']==2]
second_innings.shape
```

Data Cleaning (removing old and repeating teams)

```python
second_innings['batting_team'].value_counts()
```

```python
second_innings['batting_team']
```

```python
teams = ['Kolkata Knight Riders', 
         'Chennai Super Kings',
         'Punjab Kings', 
         'Rajasthan Royals',
         'Mumbai Indians',
         'Delhi Capitals',
         'Royal Challengers Bangalore',
         'Sunrisers Hyderabad',
         'Lucknow Super Giants',
         'Gujarat Titans'
         ]
```

```python
second_innings['batting_team'] = second_innings['batting_team'].str.replace('Delhi Daredevils','Delhi Capitals')
second_innings['bowling_team'] = second_innings['bowling_team'].str.replace('Delhi Daredevils','Delhi Capitals')

second_innings['batting_team'] = second_innings['batting_team'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
second_innings['bowling_team'] = second_innings['bowling_team'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

second_innings['batting_team'] = second_innings['batting_team'].str.replace('Kings XI Punjab','Punjab Kings')
second_innings['bowling_team'] = second_innings['bowling_team'].str.replace('Kings XI Punjab','Punjab Kings')
```

```python
second_innings['batting_team'].unique()
```

```python
second_innings = second_innings[second_innings['batting_team'].isin(teams)]
second_innings = second_innings[second_innings['bowling_team'].isin(teams)]
```

```python
second_innings.columns
```

```python
second_innings['batting_team'].unique()
```

```python
second_innings.shape
```

```python
#finding current score with cumsum method
second_innings['current_Score'] = second_innings.groupby('match_id')['total_runs_y'].cumsum()
```

```python
second_innings
```

```python
#calculate runs left after each ball using target and currenct score
second_innings['runs_left'] = second_innings['total_runs_x'] - second_innings['current_Score']
second_innings
```

```python
#calculating balls left after each ball 
second_innings['balls_left'] = 120 - (second_innings['over']*6 +second_innings['ball'])
second_innings
```

```python
second_innings
```

```python
second_innings['player_dismissed'].unique()
```

Getting Wickets

```python
#calculate the wickets left 
#filling nan values with 0 
second_innings['player_dismissed'] = second_innings['player_dismissed'].fillna("0")
```

```python
#filling where 0 is not there and there is a wicket with 1 
second_innings['player_dismissed'] = second_innings['player_dismissed'].apply(lambda x:x if x=="0" else "1")
```

```python
#converting the type to int
second_innings['player_dismissed'] = second_innings['player_dismissed'].astype('int')
```

```python
# Perform cumulative sum operation
wickets = second_innings.groupby('match_id')['player_dismissed'].cumsum()


# Calculate wickets remaining
second_innings['wickets_remaining'] = 10 - wickets
```

```python
second_innings
```

```python
second_innings['player_dismissed'].unique()
```

```python
second_innings
```

Current Run Rate

```python
second_innings['crr'] = round((second_innings['current_Score']*6)/(120 - second_innings['balls_left']),2)
second_innings
```

Required Run Rate

```python
second_innings['rrr'] = round((second_innings['runs_left']*6)/(second_innings['balls_left']),2)
second_innings
```

Finding Winner

```python
def winner(team):
    return 1 if team['batting_team'] == team['winner'] else 0
```

```python
second_innings['results'] = second_innings.apply(winner,axis=1)
second_innings
```

only taking required columns we need 

```python
required_data = second_innings[['batting_team','bowling_team','city','runs_left','balls_left','wickets_remaining','total_runs_x','crr','rrr','results']]
required_data
```

Sampling Data

```python
final_data = required_data.sample(required_data.shape[0])
```

Droping null values 

```python
final_data.dropna(inplace=True)
```

Removing rows where balls_left==0 to remove -inf and inf from rrr column

```python
final_data = final_data[final_data['balls_left']!=0]
```

```python
final_data
```

Storing data in new csv

```python
final_data.to_csv("final_data.csv", index=False)
```

<font size="5">Building Model

```python
from sklearn.model_selection import train_test_split as tts
```

```python
x = final_data.iloc[:, :-1]
y = final_data.iloc[:, -1]
```

```python
x_train, x_test, y_train, y_test = tts(x, y, test_size=0.25, random_state=100)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
# Defining a transformer for one-hot encoding categorical features
trf = ColumnTransformer([
    ('trf', OneHotEncoder(drop='first', sparse_output=False), ['batting_team', 'bowling_team', 'city'])
],
remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
# Creating a pipeline with column transformation and logistic regression
pipe = Pipeline([
    ('step1', trf),
    ('step2', LogisticRegression(solver='liblinear'))
])
```

```python
# Fitting the pipeline to the training data
pipe.fit(x_train, y_train)
```

```python
# Making predictions on the test data
y_pred = pipe.predict(x_test)
```

```python
# Calculating the accuracy of the model
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
# Predicting class probabilities for the first instance in the test set
pipe.predict_proba(x_test)[0]
```

```python
# Saving the trained pipeline as a binary file
import pickle
pickle.dump(pipe, open('pipe.pkl', 'wb'))
```
